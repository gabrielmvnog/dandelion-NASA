## Latitude and longitude assumption

The idea is to create a dataset by simulating the initial latitude and longitude of **DANDELION PROJECT** ball probes.

```python
from copy import deepcopy
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')
```

```python
# Importing the dataset
data = pd.read_csv('data.csv')
print("Input Data and Shape")
print(data.shape)
data.head()
```

```python
# Getting the values and plotting it
f1 = data['L1'].values
f2 = data['L2'].values
X = np.array(list(zip(f1, f2)))
plt.scatter(f1, f2, c='black', s=7)
```

```python
# Euclidean Distance Caculator
def dist(a, b, ax=1):
    return np.linalg.norm(a - b, axis=ax)
```

```python
# Number of clusters
k = 6
# X coordinates of random centroids
C_x = np.random.randint(0, np.max(X)-20, size=k)
# Y coordinates of random centroids
C_y = np.random.randint(0, np.max(X)-20, size=k)
C = np.array(list(zip(C_x, C_y)), dtype=np.float32)
print("Initial Centroids")
print(C)
```

```python
# Plotting along with the Centroids
plt.scatter(f1, f2, c='#050505', s=7)
plt.scatter(C_x, C_y, marker='*', s=200, c='g')
```

```python
# To store the value of centroids when it updates
C_old = np.zeros(C.shape)
```

```python
# Cluster Lables(0, 1, 2)
clusters = np.zeros(len(X))
```

```python
# Error func. - Distance between new centroids and old centroids
error = dist(C, C_old, None)
```

```python
# Loop will run till the error becomes zero
while error != 0:
    # Assigning each value to its closest cluster
    for i in range(len(X)):
        distances = dist(X[i], C)
        cluster = np.argmin(distances)
        clusters[i] = cluster
    # Storing the old centroid values
    C_old = deepcopy(C)
    # Finding the new centroids by taking the average value
    for i in range(k):
        points = [X[j] for j in range(len(X)) if clusters[j] == i]
        C[i] = np.mean(points, axis=0)
    error = dist(C, C_old, None)

colors = ['r', 'g', 'b', 'y', 'c', 'm']
fig, ax = plt.subplots()
for i in range(k):
        points = np.array([X[j] for j in range(len(X)) if clusters[j] == i])
        ax.scatter(points[:, 0], points[:, 1], s=7, c=colors[i])
ax.scatter(C[:, 0], C[:, 1], marker='*', s=200, c='#050505')
```

```python
from matplotlib import pyplot as plt
from celluloid import Camera

colors = ['r', 'g', 'b', 'y', 'c', 'm']
fig, ax = plt.subplots()
camera = Camera(fig)

def deslocar(param_inicio, param_fim):
    for p in range(k):
        points = np.array([X[j] for j in range(len(X)) if clusters[j] == p]) 
        for point in points[param_inicio:param_fim]:
            ax.scatter(point[0],point[1]+p,  c=colors[p])
            ax.scatter(point[0],point[0]+k,  c=colors[p])
            ax.scatter(point[0]+p,point[1],  c=colors[p])
            camera.snap()

lista = [[400,435],[435,455]]
for params in lista:
    deslocar(params[0], params[1])
animation = camera.animate()
for i in range(k):
    points = np.array([X[j] for j in range(len(X)) if clusters[j] == i])
    ax.scatter(points[:100, 0], points[:100, 1], s=7, c=colors[i])
animation.save('nasarentos.gif', writer = 'imagemagick')
```
